```python
from google.colab import files
uploaded = files.upload()

import os
from PIL import Image

# Load uploaded image
image_path = next(iter(uploaded))
img = Image.open(image_path).convert("RGB")
img = img.resize((512, 512))
img.save("input.png")
```

```python
import numpy as np

frames = []
num_frames = 30

for i in range(num_frames):
    zoom = 1 + (i / num_frames) * 0.3  # Slowly zoom in
    width, height = img.size
    crop_size = int(width / zoom), int(height / zoom)

    left = (width - crop_size[0]) // 2
    top = (height - crop_size[1]) // 2
    right = left + crop_size[0]
    bottom = top + crop_size[1]

    cropped = img.crop((left, top, right, bottom))
    frame = cropped.resize((width, height), Image.LANCZOS)
    frames.append(frame)
```

```python
import imageio.v3 as iio

# Save frames into a video
output_path = "zoom_video.mp4"
iio.imwrite(output_path, frames, fps=10)

# Download the video
from google.colab import files
files.download(output_path)
```
